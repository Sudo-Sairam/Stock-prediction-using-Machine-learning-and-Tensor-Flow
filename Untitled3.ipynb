```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
import tensorflow as tf
import random
import os

# Set seeds for reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed()

# Download data
ticker = 'INFY.NS'
start_date = '2018-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

data = yf.download(ticker, start=start_date, end=end_date)
close_prices = data[['Close']].values

# Scale data
scaler = MinMaxScaler()
train_data = close_prices[:-1]
scaled_train = scaler.fit_transform(train_data)
scaled_data = np.concatenate([scaled_train, scaler.transform(close_prices[-1:])])

# Prepare training data
lookback = 60
X, y = [], []
for i in range(lookback, len(scaled_data)):
    X.append(scaled_data[i-lookback:i, 0])
    y.append(scaled_data[i, 0])
X, y = np.array(X), np.array(y)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Build model
model = Sequential([
    Input(shape=(lookback, 1)),
    LSTM(50, return_sequences=True),
    LSTM(50),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, y, epochs=25, batch_size=32, shuffle=False, verbose=0)

# Predict next day's closing price
last_60_days = scaled_data[-lookback:]
X_test = last_60_days.reshape(1, lookback, 1)
predicted_scaled = model.predict(X_test, verbose=0)
predicted_price = scaler.inverse_transform(predicted_scaled)[0][0]

# Plot results
actual_prices = data['Close'].values[-lookback:]
dates = data.index[-lookback:]
predicted_plot = [None] * (lookback - 1) + [predicted_price]

plt.figure(figsize=(12, 6))
plt.plot(dates, actual_prices, label='Actual Price', linewidth=2)
plt.plot(dates, predicted_plot, label='Predicted Closing Price (Today)', color='orange', linestyle='--', marker='o')
plt.title(f'{ticker} Stock Price Prediction')
plt.xlabel("Date")
plt.ylabel("Price (‚Çπ)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f"üì¢ Predicted closing price for {datetime.today().strftime('%Y-%m-%d')}: ‚Çπ{predicted_price:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import random
import os

# Reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed()

# Load stock data
ticker = 'INFY.NS'
start_date = '2010-01-01'  # More historical data
end_date = datetime.today().strftime('%Y-%m-%d')

data = yf.download(ticker, start=start_date, end=end_date)
close_prices = data[['Close']].values

# Scaling using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(close_prices)

# Create sequences
lookback = 90
X, y = [], []
for i in range(lookback, len(scaled_data)):
    X.append(scaled_data[i-lookback:i, 0])
    y.append(scaled_data[i, 0])
X, y = np.array(X), np.array(y)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into train/validation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, shuffle=False)

# Model definition
model = Sequential([
    Input(shape=(lookback, 1)),
    LSTM(64),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train model
model.fit(X_train, y_train, validation_data=(X_val, y_val),
          epochs=100, batch_size=32, shuffle=False,
          callbacks=[early_stop], verbose=1)

# Prediction for next day
last_60_days = scaled_data[-lookback:]
X_test = last_60_days.reshape(1, lookback, 1)
predicted_scaled = model.predict(X_test, verbose=0)
predicted_price = scaler.inverse_transform(predicted_scaled)[0][0]

# Actual vs Predicted Plot
actual_prices = close_prices[-lookback:].flatten()
dates = data.index[-lookback:]
predicted_plot = [None] * (lookback - 1) + [predicted_price]

plt.figure(figsize=(12, 6))
plt.plot(dates, actual_prices, label='Actual Price', linewidth=2)
plt.plot(dates, predicted_plot, label='Predicted Closing Price (Today)', color='orange', linestyle='--', marker='o')
plt.title(f'{ticker} Stock Price Prediction')
plt.xlabel("Date")
plt.ylabel("Price (‚Çπ)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Output prediction
print(f"üì¢ Predicted closing price for {datetime.today().strftime('%Y-%m-%d')}: ‚Çπ{predicted_price:.2f}")
```
